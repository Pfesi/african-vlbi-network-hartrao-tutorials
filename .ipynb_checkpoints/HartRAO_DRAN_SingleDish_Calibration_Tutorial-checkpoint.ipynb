## Data Expectations
For a minimal drift-scan example in **CSV** form, we expect columns like:
- `time_utc` (ISO string or seconds since start)
- `elev_deg`, `az_deg` (optional; for pointing context)
- `counts` (raw backend counts)
- `cal_on` (boolean/int; 1 during noise diode ON, else 0)
- `freq_GHz` (scalar per file or column per row)
- (optional) `weather_tau`, `temp_C`, `wind_mps`

You can start from the CSV schema saved with this notebook: `drift_scan_schema_example.csv`.


```python
# === Load data ===
# Replace with your actual file path(s)
# Example: a single drift-scan CSV (one frequency)
example_file = DATA_DIR / "drift_scan_example.csv"  # <-- change me

if example_file.exists():
    df = pd.read_csv(example_file)
else:
    # Fallback: create a tiny synthetic dataset for demonstration
    t = np.arange(0, 600, 1)  # 10 minutes at 1 Hz
    # Synthetic "source" centered around t=300
    source = np.exp(-0.5*((t-300)/25)**2)
    # Raw counts with cal cycles
    cal_cycle = ((t // 30) % 2).astype(int)  # 30s cal on/off
    counts = 2000 + 80*source + 10*cal_cycle + np.random.normal(0, 3, size=t.size)
    elev = np.linspace(45, 47, t.size)  # slow elevation change
    freq_GHz = np.full_like(t, 4.8, dtype=float)

    df = pd.DataFrame({
        "time_s": t,
        "elev_deg": elev,
        "counts": counts,
        "cal_on": cal_cycle,
        "freq_GHz": freq_GHz
    })

print(df.head())
print(df.describe())
```

## 1) Inspect Raw Counts
Look at overall trends, cal-on/off plateaus, and the source peak (transit in a drift scan).

```python
plt.figure(figsize=(10,4))
plt.plot(df.get("time_utc", df["time_s"]), df["counts"], lw=1)
plt.title("Raw Counts vs Time")
plt.xlabel("Time (UTC or s)")
plt.ylabel("Counts")
plt.tight_layout()
plt.show()
```

## 2) Compute System Temperature (Tsys)
If you have a known **noise diode equivalent temperature** \(T_{cal}\) and can measure the **power difference** between cal-on and cal-off plateaus, you can estimate \(T_{sys}\).

A common radiometer relation using the diode step is:
\[
Y = \frac{P_{\mathrm{on}}}{P_{\mathrm{off}}}, \qquad T_{\mathrm{sys}} \approx \frac{T_{cal}}{Y - 1}
\]

> **Note:** In practice, compute \(P_{\mathrm{on}}\) and \(P_{\mathrm{off}}\) as robust averages over stable windows (exclude the source!).

```python
# --- Identify cal-on/off samples away from the source ---
# For a quick demo, we classify source-free regions by thresholding a smoothed version.
from scipy.ndimage import gaussian_filter1d

smooth_counts = gaussian_filter1d(df["counts"].values, sigma=10)
peak_mask = smooth_counts > (np.median(smooth_counts) + 3*np.std(smooth_counts))  # naive mask for source
bg_mask = ~peak_mask

cal_on_mask = (df["cal_on"] == 1) & bg_mask
cal_off_mask = (df["cal_on"] == 0) & bg_mask

# Robust plateaus
P_on = np.median(df.loc[cal_on_mask, "counts"]) if cal_on_mask.any() else np.nan
P_off = np.median(df.loc[cal_off_mask, "counts"]) if cal_off_mask.any() else np.nan

# User-supplied T_cal (K) — replace with your receiver's noise diode temp (per band!)
T_cal_K = 5.0  # <-- change me per receiver/frequency

Y = P_on / P_off if (np.isfinite(P_on) and np.isfinite(P_off) and P_off != 0) else np.nan
Tsys_K = T_cal_K / (Y - 1) if (np.isfinite(Y) and (Y - 1) != 0) else np.nan

print(f"P_on={P_on:.2f}, P_off={P_off:.2f}, Y={Y:.5f}, Tsys≈{Tsys_K:.2f} K")
```

## 3) Extract Antenna Temperature \(T_A^*\)
Convert raw counts to \(T_A^*\) using the cal step:
\[
\Delta P = P_{\mathrm{on}} - P_{\mathrm{off}} \quad\Rightarrow\quad \frac{T_{cal}}{\Delta P} \; \text{[K / count]}
\]
Then apply to the **source** portion of the scan (after baseline removal).

```python
# K-per-count scale from the diode step
dP = P_on - P_off if (np.isfinite(P_on) and np.isfinite(P_off)) else np.nan
K_per_count = T_cal_K / dP if (np.isfinite(dP) and dP != 0) else np.nan

# Quick baseline removal using a polynomial fit on background (no source)
x = df.get("time_utc", df["time_s"]).values
y = df["counts"].values

# Fit on background only
xb = x[bg_mask]
yb = y[bg_mask]
coefs = np.polyfit(xb, yb, deg=2)  # start with a low-order polynomial
baseline = np.polyval(coefs, x)

resid = y - baseline  # baseline-subtracted counts
Ta_star_K = resid * K_per_count   # antenna temperature scale

df["baseline"] = baseline
df["resid_counts"] = resid
df["Ta_star_K"] = Ta_star_K

print(f"K_per_count ≈ {K_per_count:.5f} K/count")
```

```python
plt.figure(figsize=(10,4))
plt.plot(x, y, label="Raw")
plt.plot(x, baseline, label="Baseline (polyfit)")
plt.title("Raw Counts and Baseline")
plt.xlabel("Time")
plt.ylabel("Counts")
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,4))
plt.plot(x, df["resid_counts"], label="Residual (Counts)")
plt.title("Baseline-Subtracted Residual")
plt.xlabel("Time")
plt.ylabel("Counts")
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,4))
plt.plot(x, df["Ta_star_K"], label="Antenna Temperature $T_A^*$")
plt.title("Antenna Temperature vs Time")
plt.xlabel("Time")
plt.ylabel("K")
plt.legend()
plt.tight_layout()
plt.show()
```

## 4) Compute Gain (K/Jy) from a Flux Calibrator
Observe a standard calibrator with known **flux density** \(S_{cal}\) at the observing frequency (e.g., Hydra A, 3C286).  
If the peak (or integrated) **\(T_A^*\)** on the calibrator is measured, then:
\[
G \; [\mathrm{K/Jy}] = \frac{T_{A,cal}^*}{S_{cal}}
\]
and thus for any target,
\[
S_{\mathrm{target}} \; [\mathrm{Jy}] = \frac{T_{A,target}^*}{G}
\]

```python
# --- Calibrator-based gain ---
# Replace with your calibrator's known flux at this frequency (from Perley & Butler, or your site table)
S_cal_Jy = 43.0  # e.g., Hydra A at ~4.8 GHz (illustrative; use your proper value)

# Estimate T_A* for calibrator (peak or integrated). Here we'll use peak of our synthetic scan.
Ta_cal_peak_K = np.nanmax(df["Ta_star_K"].values)
G_K_per_Jy = Ta_cal_peak_K / S_cal_Jy if S_cal_Jy != 0 else np.nan

print(f"Estimated gain G ≈ {G_K_per_Jy:.5f} K/Jy (from calibrator)")
```

## 5) Apply Calibration to Convert \(T_A^*\) → Flux Density (Jy)
\[
S_{\mathrm{Jy}} = \frac{T_A^*}{G}
\]

```python
df["Flux_Jy"] = df["Ta_star_K"] / G_K_per_Jy

plt.figure(figsize=(10,4))
plt.plot(x, df["Flux_Jy"])
plt.title("Calibrated Flux Density vs Time")
plt.xlabel("Time")
plt.ylabel("Flux Density (Jy)")
plt.tight_layout()
plt.show()

# Save calibrated time series
out_csv = OUTPUT_DIR / "calibrated_timeseries.csv"
df.to_csv(out_csv, index=False)
print("Saved:", out_csv)
```

## 6) Quality Control & Checks
- **Pointing:** Compare beam profile vs expected FWHM; large offsets reduce peak.  
- **Weather:** Look for rapid Tsys swings or attenuation with elevation.  
- **RFI:** Spiky residuals or narrowband contamination (for total power, often bursts).  
- **Gain stability:** Compare G across days/frequencies.  
- **Baseline:** Try different polynomial orders; avoid overfitting the source region.


### Beam Profile and FWHM (Quick Check)
For a Gaussian beam with FWHM \(\\theta_{\\mathrm{FWHM}}\), the response vs offset \(\\theta\) is:
\[
B(\\theta) = \\exp\\Big(-4\\ln 2\\ (\\theta/\\theta_{\\mathrm{FWHM}})^2\\Big).
\]
If you have an **az/el scan**, fit a Gaussian to the source crossing to estimate the observed FWHM and compare with the theoretical \(\\sim 1.22\\lambda/D\\) for a circular aperture (in radians).

```python
# Optional: Fit a Gaussian to estimate observed FWHM (requires a clean cross-scan)
# Here we demonstrate on time axis as a proxy if the drift approximates a Gaussian crossing.
from scipy.optimize import curve_fit

def gauss(x, A, x0, sigma, C):
    return A*np.exp(-(x-x0)**2 / (2*sigma**2)) + C

# Use the antenna temperature as signal
y_sig = df["Ta_star_K"].values

# Initial guesses
A0 = y_sig.max() - np.median(y_sig)
x0 = x[np.argmax(y_sig)]
sigma0 = (x.max()-x.min())/50
C0 = np.median(y_sig)

try:
    popt, pcov = curve_fit(gauss, x, y_sig, p0=[A0, x0, sigma0, C0])
    A, x0_fit, sigma_fit, C = popt
    fwhm_est = 2*np.sqrt(2*np.log(2))*sigma_fit  # in 'x' units
    print(f"Gaussian fit: peak≈{A+C:.3f} K, center≈{x0_fit:.2f}, sigma≈{sigma_fit:.3f}, FWHM≈{fwhm_est:.3f} (x-units)")
    
    xx = np.linspace(x.min(), x.max(), 1000)
    yy = gauss(xx, *popt)

    plt.figure(figsize=(10,4))
    plt.plot(x, y_sig, '.', ms=3, label="Data")
    plt.plot(xx, yy, lw=2, label="Gaussian fit")
    plt.title("Gaussian Fit to Beam Crossing (Proxy)")
    plt.xlabel("Offset (proxy: time)")
    plt.ylabel("Antenna Temperature (K)")
    plt.legend()
    plt.tight_layout()
    plt.show()
except Exception as e:
    print("Gaussian fit skipped/failed:", e)
```
